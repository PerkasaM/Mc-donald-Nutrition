```python

df_num = result.select_dtypes(include='number')

# Determine the number of rows and columns for the subplot grid
num_rows = (len(df_num.columns) - 1) // 3 + 1
num_cols = min(len(df_num.columns), 3)

# Create subplot grid
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 4 * num_rows))

# Loop through each numeric column and create a boxplot
for i, column in enumerate(df_num.columns):
    row_index = i // num_cols
    col_index = i % num_cols
    sns.boxplot(ax=axes[row_index, col_index], x=df_num[column])
    axes[row_index, col_index].set_title(f'Boxplot for {column}')

plt.tight_layout()
plt.show()
```

### Let's study how many calories each food category contains. Therefore helping the health conscious people to select the perfect combination :-)

```python
# Getting total number of calories for each food item in a separate column
result['Total Calories']=result['Calories']+result['Calories from Fat']
```

```python
result['Total Calories'].head()
```

```python
# Rounding Off the calories to nearest hundred so that data can be handled and analyzed easily
def roundOff(x):
    if x==0:
        x=0
    elif x<100:
        x=50
    else:
        y=x%100
        if y<50:
            x=x-y
        else:
            x=x-y+100
    return x

result['Estimated Calories']=result['Total Calories'].apply(roundOff)
result['Estimated Calories'][:5]
```

```python
# Mengelompokkan data dan mengambil nilai terbanyak untuk setiap kategori
x1 = result.groupby('Category').agg(lambda x: x.value_counts().index[0])

# Menambahkan sort_values() untuk mengurutkan DataFrame berdasarkan nilai Estimated Calories
x1 = x1.sort_values(by='Estimated Calories', ascending=False)

# Membuat plot bar menggunakan palet 'inferno'
sns.barplot(palette='inferno', x=x1.index, y=x1['Estimated Calories'].values)
plt.xticks(rotation=90)
plt.show()
```

 From the above graph its depicted that food items in 'Breakfast','Chicken & Fish' and 'Smoothies and Shakes' category contains the maximum amount of calories whereas amount of calories in 'Snacks & Sides','Salads' and 'Beef & Pork' are in a moderate amount while the lowest calories are in the food items that covers the 'Coffee & Tea' and 'Deserts' category.
